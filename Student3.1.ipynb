```python
import tkinter as tk
from tkinter import ttk

# List to store student names
student_list = []

# Function to add student
def add_student():
    name = name_entry.get().strip()
    if name:
        student_list.append(name)
        update_display()
        name_entry.delete(0, tk.END)

# Function to remove selected student
def remove_student():
    selected = student_listbox.curselection()
    if selected:
        del student_list[selected[0]]
        update_display()

# Update the list display
def update_display():
    student_listbox.delete(0, tk.END)
    for student in student_list:
        student_listbox.insert(tk.END, student)

# GUI window
root = tk.Tk()
root.title("üìò Student Registration System")
root.geometry("450x500")
root.configure(bg="#e6f2ff")

style = ttk.Style()
style.theme_use("clam")

# Custom style
style.configure("TButton",
                font=("Arial", 12),
                padding=6,
                background="#3399ff",
                foreground="white")
style.map("TButton",
          background=[('active', '#1a75ff')])

# Title label
title = tk.Label(root, text="üìù Student Name Entry", font=("Helvetica", 18, "bold"), bg="#e6f2ff", fg="#003366")
title.pack(pady=20)

# Entry field
name_entry = ttk.Entry(root, font=("Arial", 14), width=30)
name_entry.pack(pady=10)

# Add button
add_btn = ttk.Button(root, text="‚ûï Add Student", command=add_student)
add_btn.pack(pady=5)

# List label
list_label = tk.Label(root, text="üéì Registered Students:", font=("Arial", 12, "bold"), bg="#e6f2ff", fg="#003366")
list_label.pack(pady=10)

# Listbox
student_listbox = tk.Listbox(root, height=10, width=40, font=("Arial", 12), bd=0, highlightthickness=1, selectbackground="#cce0ff")
student_listbox.pack(pady=5)

# Remove button
remove_btn = ttk.Button(root, text="‚ùå Remove Selected", command=remove_student)
remove_btn.pack(pady=15)

# Footer
footer = tk.Label(root, text="Built with ‚ù§ in Python", bg="#e6f2ff", fg="gray", font=("Arial", 10))
footer.pack(side="bottom", pady=10)

root.mainloop()
```
